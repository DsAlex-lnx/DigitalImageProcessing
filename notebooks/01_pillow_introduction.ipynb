# Introdução ao Processamento Digital de Images

### O Que É Uma Imagem Digital?



Na era moderna, a imagem digital emerge como uma tradução da realidade para uma grade de pixels — minúsculos pontos de luz que, organizados, formam o todo. 

Diferente das imagens físicas, a imagem digital é maleável, composta de valores numéricos, e permite uma ampla gama de manipulações e criações. 

Trabalhar com imagens digitais é, portanto, uma jornada que combina arte e tecnologia...

>Objetivos: entender imagens digitais e aprender a criar e manipular imagens com a biblioteca Pillow.


### O Que É Um Pixel?

Um **pixel** (abreviação de "picture element") é a menor unidade de uma imagem digital, representando um único ponto de cor em uma grade organizada. 
- Cada pixel contém informações de cor e, quando combinados em uma matriz, formam a imagem completa que vemos na tela. 
- Quanto mais pixels uma imagem possui, maior é sua resolução, resultando em uma imagem mais detalhada.

Na imagem abaixo, cada quadradinho representa um pixel individual. Juntos, esses pixels compõem uma imagem de "pixel art". 

Esse estilo artístico é uma ótima forma de visualizar como as imagens digitais são construídas a partir de uma organização de pequenos elementos coloridos.

### Conceitos Fundamentais de Imagens Digitais



Ao trabalhar com imagens digitais, alguns conceitos são essenciais para entender como elas são criadas e manipuladas. 

---

#### Resolução e Dimensões de Uma Imagem
- **Resolução**: Refere-se ao número de pixels em uma imagem. É frequentemente expressa em DPI (pontos por polegada) ou pela contagem de pixels (por exemplo, 1920x1080).
- **Dimensões**: Medem a largura e a altura de uma imagem em pixels. Dimensões maiores indicam uma imagem com mais detalhes, mas também podem aumentar o tamanho do arquivo.

#### Modelos de Cores
- **RGB (Red, Green, Blue)**: Modelo usado em telas, onde cada pixel é uma combinação de vermelho, verde e azul.
- **Escala de Cinza**: Usa apenas tons de cinza, variando do preto ao branco, sendo ideal para imagens sem cor.
- **CMYK (Cyan, Magenta, Yellow, Black)**: Modelo utilizado para impressão, combinando cores de tinta para formar a imagem final no papel.

#### Formatos de Imagens Comuns
- **JPEG**: Formato comum com compressão com perdas, ideal para fotografias e imagens coloridas.
- **PNG**: Suporta transparência e usa compressão sem perdas, adequado para gráficos e logotipos.
- **BMP**: Formato sem compressão, resultando em arquivos grandes, mas de alta qualidade.
- **GIF**: Suporta animações e cores limitadas, usado em imagens de baixa resolução ou pequenas animações.

## Exemplo Prático: Bandeira do Japão

### Características da Bandeira Japonesa


Neste exemplo, vamos criar a bandeira do Japão utilizando a biblioteca Pillow. A bandeira é composta por um fundo branco e um círculo vermelho centralizado, representando o sol nascente. 

<img src='../resources/bandeira_japao.png' width='800'>

- **Proporção**: A bandeira do Japão segue uma proporção de 2:3, ou seja, a largura é 1,5 vezes a altura.
- **Cores**:
  - **Fundo**: Branco (RGB: 255, 255, 255)
  - **Círculo Central**: Vermelho (RGB: 173, 35, 51) — padrão de cor RGB para o vermelho utilizado.
- **Dimensões**:
  - **Altura**: 500 pixels
  - **Largura**: 750 pixels (calculada com base na proporção de 2:3)
  - **Raio do Círculo**: 150 pixels (3/10 da altura da bandeira)
- **Centro do Círculo**: Localizado no ponto (375, 250) da imagem, que é o centro da bandeira.

### Equação do Círculo

A equação padrão de um círculo no plano cartesiano é:

**x² + y² = r²**

Onde:
- **(0, 0)** é o centro do círculo.
- **r** é o raio do círculo.

<img src='../resources/equacao_circulo_01.webp' width='300'>

#### Expansão para a Equação Geral

A equação geral de um círculo no plano cartesiano é dada por:

**x² + y² - 2ax - 2by + a² + b² - r² = 0**

<img src='../resources/equacao_circulo_02.webp' width='300'>

Onde:
- **(a, b)** são as coordenadas do centro do círculo.
- **r** é o raio do círculo.

#### Reorganizando para a Forma Padrão

Podemos rearranjar a equação geral para a forma padrão do círculo:

1. **Reorganizando a equação**:  
   Para transformar a equação geral em sua forma padrão, isolamos os termos de (x, y):

   **(x - a)² + (y - b)² = r²**

   Onde:
   - **(x - a)²** representa a parte da equação associada ao eixo x.
   - **(y - b)²** representa a parte da equação associada ao eixo y.

#### Cálculo do Raio do Círculo

A altura da imagem é de **H** pixels, e o círculo representa uma fração **F** da altura da bandeira. Para calcular o tamanho do círculo, usamos a seguinte fórmula:

$$
\text{Tamanho do Círculo} = H \times F
$$

O raio do círculo é, então, a metade do tamanho do círculo:

$$
\text{Raio} = \frac{\text{Tamanho do Círculo}}{2} = \frac{H \times F}{2}
$$

Portanto, o cálculo `radius = F * height // 2` garante que o círculo central tenha o tamanho apropriado em relação à altura da bandeira.

```python
WHITE = (255, 255, 255)
RED = (173, 35, 51)

height = 500
width = 3*height//2

japan = Image.new('RGB', (width, height), WHITE)

radius = int ((3/5)*height)//2
center = (width//2, height//2)

for x in range(center[0]-radius, center[0]+radius):
    for y in range(center[1]-radius, center[1]+radius):
        if ((x-center[0])**2 + (y-center[1])**2) <= radius**2:
            japan.putpixel((x, y), RED)

japan
```
